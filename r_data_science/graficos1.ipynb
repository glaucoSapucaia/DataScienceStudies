## Db trees

```r
summary(trees)
head(trees)
```

## Argumentos
- main= -> Titulo principal
- xlab= | ylab= -> Nome dos eixos
- col= -> Cores
- density= -> Densidade das cores
- breaks= -> Intervalos entre os dados analisados

```r
hist(trees$Girth, main='Cincunferências', xlab = 'Circunferências',
ylab='Frequência', col='blue', density=300, breaks=4)
```

## Gráficos de densidade

```r
densidade = density(trees$Girth)
plot(densidade)
```

## Sobrepondo gráficos
- Valores NULL para evitar a sobreposição de titulos!

## A função par() cria sobreposições!

```r
hist(trees$Girth, main=NULL, xlab=NULL, ylab=NULL, breaks=4)
par(new=T)
plot(densidade)
```

## Graficos de dispersão!
- Ideal para comparação de duas variaveis continuas e suas relações!
- pch= -> Altera o elemento gráfico | Veja a tabela de indices.
- type= -> Também os pontos de marcação. 'l' = linha continua!

```r
plot(trees$Girth, trees$Volume, main='Árvores', xlab='Circunfêrencia',
ylab='Volume', col='purple', pch=19, type='l')
```

## jitter() -> Altera sobreposições

```r
plot(jitter(trees$Girth), trees$Volume, main='Árvores', xlab='Circunferência',
ylab='Volume', col='red', pch=19)
```

## Indentificando dados por cores e legendas
- legend() -> Adiciona legendas
- x= -> Posição da legenda
- legend= -> Nomes da legenda | Indique os dados na ordem em que aparecem no
banco de dados.
- fill= -> Indica as cores para a legenda
- cex= -> Tamanho da caixa de legenda

```r
plot(CO2$conc, CO2$uptake, col=CO2$Treatment, pch=19)
legend(x='bottomright', legend=c('nonchilled', 'chilled'),
fill=c('black', 'red'), cex=1)
```

## Agrupando gráficos! | dashboards
- As posição das colunas indicam os eixos | x, y no gráfico

```r
plot(trees)
boxplot(trees, horizontal=F, col='purple')
```
