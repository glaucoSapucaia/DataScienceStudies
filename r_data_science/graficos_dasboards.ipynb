## Argumentos
- main= -> Titulo principal
- xlab= | ylab= -> Nome dos eixos
- col= -> Cores
- density= -> Densidade das cores
- breaks= -> Intervalos entre os dados analisados

```r
hist(trees$Girth, main='Cincunferências', xlab = 'Circunferências',
ylab='Frequência', col='blue', density=300, breaks=4)
```

## Gráficos de densidade

```r
densidade = density(trees$Girth)
plot(densidade)
```

## Sobrepondo gráficos
- Valores NULL para evitar a sobreposição de titulos!

## A função par() cria sobreposições!

```r
hist(trees$Girth, main=NULL, xlab=NULL, ylab=NULL, breaks=4)
par(new=T)
plot(densidade)
```

## Graficos de dispersão!
- Ideal para comparação de duas variaveis continuas e suas relações!
- pch= -> Altera o elemento gráfico | Veja a tabela de indices.

```r
plot(trees$Girth, trees$Volume, main='Árvores', xlab='Circunfêrencia',
ylab='Volume', col='purple', pch=19)
```
