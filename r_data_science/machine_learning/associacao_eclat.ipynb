```r
library(arules)
library(arulesViz)

data_ = read.transactions("transacoes.txt", sep = ",", format = "basket")
data_
image(data_)
```

## Regras
- eclat( ) -> Retorna regras de associação. Só é necessário indicar o tamanho minimo da regra

```r
regras = eclat(data_, parameter = list(minlen=2))
inspect(regras)
```

## Visualização Gráfica

```r
# Tamanho da circunferência = Suporte
plot(regras, method='graph', engine='igraph')
```
