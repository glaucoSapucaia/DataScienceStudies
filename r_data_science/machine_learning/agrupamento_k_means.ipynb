## Packages
- install.packages('factoextra')

```r
library(factoextra)

dim(iris)
head(iris)
summary(iris)
```

## Cluster (agrupamentos)
- kmeans( ) -> Retorna agrupamento (cluster) dos dados
- centers= -> Indica a quantidades de clusters
- kmeans agrupa todos os elementos - Não gera ruídos

```r
cluster_ = kmeans(iris[1:4], centers = 3)
cluster_

# Criamos uma tablea para comparar o cluster as espécies
table(iris$Species, cluster_$cluster)
```

## Gráficos Clusters

```r
plot(iris$Sepal.Length, iris$Sepal.Width, col=cluster_$cluster, pch=19, main='Iris Cluster Sepal')
plot(iris[, 1:4], col=cluster_$cluster, pch=19, main='Iris Cluster Sepal and Nepal')
```

## Gráfico Especializado

```r
g2 = fviz_cluster(cluster_, data=iris[1:4], ellipse.type = 'convex',
ggtheme = theme_bw(), main='Cluster')
plot(g2)
```

## Adicionando Cluster aos dados originais
- write.csv( ) -> Salavamos a nova base de dados

```r
iris2 = iris
iris2['cluster'] = cluster_$cluster

dim(iris2)
iris2
summary(iris2)

write.csv(iris2, 'iris2.csv')
```
