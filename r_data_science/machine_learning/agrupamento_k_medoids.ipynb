## Packages
- install.packages('cluster')

```r
library(factoextra)
library(cluster)

dim(iris)
head(iris)
summary(iris)
```

## Cluster (agrupamentos)
- pam( ) -> Retorna agrupamento (cluster) dos dados
- k= -> Indica a quantidades de clusters. Inicia em pontos indicado pelos dados
- kmedoids agrupa todos os elementos POSSÍVEIS - Pode gerar ruídos

```r
cluster_ = pam(iris[, 1:4], k=3)
cluster_

# Visualização gráfica
plot(cluster_)

# Criação de tabela para comparação dos clusters com as espécies
table(iris$Species, cluster_$clustering)
```

## Gráfico com fviz_cluster( )

```r
g = fviz_cluster(list(data=iris[, 1:4], cluster=cluster_$clustering),
ellipse.type = 'norm', ggtheme = theme_bw(), main='Iris Cluster Medoids')
plot(g)
```
