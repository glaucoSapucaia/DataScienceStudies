## Packages
- install.packages('gridExtra')

```r
library(factoextra)
library(gridExtra)

dim(iris)
head(iris)
summary(iris)
```

## CLuster Hierárquico
- hcut( ) -> Retorna os agrupamentos
- hc_method= -> Define o método de cluster
- stand= -> Indica a padronização dos dados

```r
hc = hcut(iris[, -5], k=3, hc_method = 'complete', stand=T)
hc
```

## Dendogramas
- fviz_dend( ) -> Retorna um dendograma
- show_labels= -> Indica a visualização dos labels
- rect= -> Indica a criação de retangulos para observação

```r
plot(hc, cex=0.6)

g1 = fviz_dend(hc, show_labels = F, rect=T, ggtheme = theme_bw(),
main='Cluster Hierárquico | Iris')
plot(g1)
```

## Gráfico Cluster 'Padrão'

```r
g2 = fviz_cluster(hc, ellipse.type = 'norm', ggtheme = theme_bw(),
main='Iris CLuster')
plot(g2)
```
