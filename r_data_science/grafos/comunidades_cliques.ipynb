## DB
- as.undirected() -> tranforma num grafo n√£o direcionado

```r
library(igraph)

grafo = read_graph("Grafo.graphml", format = c("graphml"))
grafo = as.undirected(grafo)
plot(grafo)
```

## Comunidades
- cluster_edge_betweenness() - > Retorna um vetor de comunidades detectadas

```r
comun = cluster_edge_betweenness(grafo)
comun

# Retorna aas comunidades geradas
comun$membership

# Diferenciando por cores
plot(grafo, vertex.color=comun$membership)
```

## Mais testes de comunidades

```r
grafo2 = graph(
    edges = c(
        "A", "C", "A", "B", "B", "E", "B", "F", "C", "D", "G", "H", "D",
        "H", "E", "H", "F", "G"
    )
)
comun2 = cluster_edge_betweenness(grafo2)
comun2
comun2$membership
plot(grafo2, vertex.color=comun2$membership, edge.color='gray')
```

## Cliques
- cliques() -> Retorna cliques detectados | Apenas grafo n√£o direcionados
- min= -> Numero minimo de menbros para um clique

```r
clique_ = cliques(as.undirected(grafo2), min=2)

# Qtd de cliques
length(clique_)

#Detalhes de cliques
clique_
```
