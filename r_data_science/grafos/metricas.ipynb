```r
library(igraph)

grafo = read_graph('Grafo.graphml', format=c('graphml'))
grafo
plot(grafo)
```

## Graus de vétice (Node)
- degree( ) -> Retorna grau especificado
- mode= -> all=entrada e saida, in=entrada, out=saida

```r
# graus
degree(grafo, mode='all')
degree(grafo, mode='in')
degree(grafo, mode='out')

# Indicando tamanho do vertice de acordo com seu grau
grau = degree(grafo, mode='in')
plot(grafo, vertex.size=grau)
```

## Diametros | maior distancia entre vértices
- diameter() -> Retorna a maior distancia
- get_diameter() -> Retorna o percurso do edge
- directed= -> Indica se considera edges direcionados ou nao

```r
# Maior distancia com edges direcionados
diameter(grafo, directed = T)
get_diameter(grafo, directed = T)

# Maior distancia sem edges direcionados
diameter(grafo, directed = F)
get_diameter(grafo, directed = F)
```

## Vizinhança | Densidade | Isomorfismo
- neigborhood( ) -> Retorta a vizinhança (conexões de um vertice)
- 0 -> Este valor indica para considerar o próprio vértice do inicio
- mode= -> 'all' = Indica a analise de edges sem considerar direção
- 
- edge_density() -> Retorna densidade do grafo | Relação entre arestas e vertices
- isomorphic() -> Retorna se ops grafos comparado são isomorficos (semelhantes, mesma estrutura)

```r
neighborhood(grafo, 0, mode='all')
edge_density(grafo)

grafo2 = grafo
grafo3 = graph_from_literal(1-+2-3-4-5+-4+-1-+3)
plot(grafo3)

isomorphic(grafo, grafo2)
isomorphic(grafo, grafo3)
```
