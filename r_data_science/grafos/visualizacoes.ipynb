```r
library(igraph)
library(visNetwork)

grafo = graph(edges = c("ana", "maria", "fernando", "jose", "ana", "fernando", "ana", "jose"))
V(grafo)$Peso = c(40, 30, 30, 25)
E(grafo)$TipoAmizade = c("amigo", "inimigo", "inimigo", "amigo")
E(grafo)$weight = c(1, 2, 1, 3)
V(grafo)$type = "Humanos"
grafo
plot(grafo)
```

## Impressões
- vertex.size= -> Retorna tamanho do vértice de acordo com 'peso' do atributo
- edge.width= -> Retorna tamanho da aresta de acordo com 'peso' do atributo

```r
# Tamanho do vértice
plot(grafo, vertex.size=vertex_attr(grafo)$Peso)

# Tamanho da aresta
plot(grafo, vertex.sizw=vertex_attr(grafo)$Peso, edge.width=edge_attr(grafo)$weight)
```

## Cores
- Utilizamos cores como atributos
- vertex.color -> retorna cores

```r
vertex_attr(grafo)$Cor = c('Blue', 'Red', 'Yellow', 'Green')
plot(
    grafo, vertex.size=vertex_attr(grafo)$Peso,
    edge.width=edge_attr(grafo)$weight,
    vertex.color=vertex_attr(grafo)$Cor
)
```

## Curvatura, frame e título
- edge.curved= -> Indica curvatura de arestas
- frame= -> Cria uma borda no grafo
- main= -> Define título
- vertex.shape= -> Indica forma do vértice

```r
plot(
    grafo, vertex.size=vertex_attr(grafo)$Peso,
    vertex.color=vertex_attr(grafo)$Cor,
    edge.width=edge_attr(grafo)$weight,
    edge.curved=0.4,
    frame=T,
    main='Grafo'
)

plot(
    grafo, vertex.size=vertex_attr(grafo)$Peso,
    vertex.color=vertex_attr(grafo)$Cor,
    edge.width=edge_attr(grafo)$weight,
    edge.curved=0.4,
    frame=T,
    main='Grafo',
    vertex.shape='square'
)
```
