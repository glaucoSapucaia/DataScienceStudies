```r
library(igraph)

grafo = graph(
    edges = c(
        "A", "C", "A", "B", "B", "E", "B", "F", "C", "D", "G", "H", "D",
        "H", "E", "H", "F", "G"
    )
)
plot(grafo)
```

## Definindo Pesos
- edge.label= -> Define o nome de edges

```r
E(grafo)$weight = c(2, 1, 2, 1, 2, 1, 1, 3, 1)
plot(grafo, edge.label=E(grafo)$weight)
```

## Calculando Distancias
- distances() -> Retorna a distancia mais curta entre nodes especificos, NÂO mostra
- shortest_paths() -> Retorna a distancia mais curta, MOSTRA
- output= -> Indica se resultado retorne Nodes, Edges ou both

```r
distances(grafo, V(grafo)$name == 'A', V(grafo)$name == 'H')
plot(grafo, edge.label=E(grafo)$weight)

caminho = shortest_paths(grafo, V(grafo)$name == 'A', V(grafo)$name == 'H',
output = 'both',
)
caminho
caminho$vpath

# Demonstrando caminho mais curto no gráfico
# Diferanciação com cores

for (i in 1:length(V(grafo))) {
   V(grafo)$color[i] = ifelse(i %in% as.vector(unlist(caminho$vpath)),
   'green', 'gray')
}

for (i in 1:length(E(grafo))) {
    E(grafo)$color[i] = ifelse(i %in% as.vector(unlist(caminho$epath)),
    'green', 'gray')
}

plot(grafo, edge.label=E(grafo)$weight)
```
