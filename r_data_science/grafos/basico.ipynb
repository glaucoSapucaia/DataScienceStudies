## Edges Literais e grafos nÃ£o direcionados
- graph_from_literal( ) -> Retorna um grafo
- \+ -> Indica a direÃ§Ã£o da aresta
- \- -> Indica o inicio da aresta
- apenas \- -> Indica um grafo nÃ£o direcionado

```r
grafo5 = graph_from_literal(1-+2, 2-+3, 3++4, 4+-1)
plot(grafo5)

grafo6 = graph_from_literal(1-2, 2-3, 3-4, 4-1)
plot(grafo6)
```

## Grafo desconexo | vÃ©rtice isolado | grau zero
- n= -> Completa com nodes isolados atÃ© o valor indicado
- isolates= -> Retorna nodes isolados

```r
grafo7 = graph_from_literal(1-2-3-4-1-3, 5)
plot(grafo7)

grafo10 = graph(edges = c(1, 2, 2, 3, 3, 4, 4, 1), directed = F, n=10)
grafo10
plot(grafo10)

grafo12 = graph(edges = c('a', 'b', 'b', 'c', 'c', 'a'), isolates = c('I', 'j'))
plot(grafo12)
```

## Analisando Grafo
### Return
- D/U -> direcionado ou nÃ£o-direcionado
- N -> Nome do grafo
- B -> Tipo do grafo
- W -> grafo ponderado (peso em arestas)
- DimensÃµes vÃ©rtices | arestas (4 5)
- directed= -> Indica grafo direcionado ou nao
- 
- IGRAPH 5833362 D--- 4 5 --
- \+ edges from 5833362:
- [1] 1->2 2->3 3->4 4->1 1->3

```r
#direcionado
grafo8 = graph(edges = c(1, 2, 2, 3, 3, 4, 4, 1, 1, 3))
grafo8
plot(grafo8)

# nÃ£o direcionado
grafo9 = graph(edges = c(1, 2, 2, 3, 3, 4, 4, 1), directed = F)
grafo9
plot(grafo9)
```

## Mais ATRIBUTOS
- vertex_attr( ) -> Retorna atributos de Nodes
- edge_attr( ) -> Retorna attrs de Edges
- E( ) -> Usado para influenciar arestas
- $weight -> Peso padrÃ£o de edges | grafo ponderado
- $type -> Tipo do grafo

```r
grafo13 = graph(edges = c('ana', 'lucas', 'lucas', 'joao', 'joao', 'ana'))

# criando atributos para Nodes
V(grafo13)$Peso = c(40, 30, 25)

# criando attrs para edges
E(grafo13)$TipoAmizade = c('amigo', 'amigo', 'inimigo')

# Observando attrs
vertex_attr(grafo13)
edge_attr(grafo13)

# selecionando attrs
V(grafo13)$Peso
E(grafo13)$TipoAmizade

# Peso padrÃ£o de edges
E(grafo13)$weight = c(1, 2, 2)

# Tipo do grafo
V(grafo13)$type = 'Humanos'

grafo13

plot(grafo13)
```
