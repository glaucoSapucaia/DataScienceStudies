## Package -> forecast
- install.packages('forecast')

## DB -> AirPassengers
- É uma série temporal
- start() -> Inicio da série temporal
- end() -> Fim da série temporal
- frequency() -> Frequencia (tempo) dos dados (Em AirPassengers | Anos)
- Ainda sobre frequency() -> Retorna 12 intervalos, interpretados como meses, temos os dados por ano.

```r
library(forecast)

AirPassengers
start(AirPassengers)
end(AirPassengers)
frequency(AirPassengers)
```

## Gráficos
- intervalo -> eixo x
- dados (valor) -> eixo y
- window() -> Retorna uma janela (periodo) especifico dos dados

```r
# Retorna a série temporal
plot(AirPassengers)

# Agregando valores para observar tendencia
plot(aggregate(AirPassengers))

# Janela (parte) dos dados
part = window(AirPassengers, start=c(1960, 1), end=c(1960, 12))
plot(part)
```

## Decomposição
- retornos
- $x -> Dados originais
- $seasonal -> Elementos sazionais
- $trend -> Elementos de tendencia
- $random -> Elementos de aleatoriedade
- attributes() -> retorna tipos de elementos encontrados
- $ -> Modo geral do R para selecionar dados

```r
dec = decompose(AirPassengers)
dec

# retorna atributos encontrados
attributes(dec)

# Retorna apenas o atributo indicado
dec$seasonal
```

## Decomposição (GRÁFICA)

```r
# Retorna x, seasonal, trend e random
plot(dec)

# Elementos isolados
plot(dec$x, col='blue')
plot(dec$seasonal, col='red')
plot(dec$trend, col='purple')
plot(dec$random, col='green')
```

## Previsões - forecast (ets)
- ets() -> Realiza suavização exponencial (dados mais recentes na linha do tempo, possuem mais relevancia para a previsão)
- forecast() -> Retorna a previsão | Recebe o modelo e h= -> intervalos a frente para a previsão | 12 igual a proximo ano

```r
# Suavização exponencial
# Atenção aos indices - AIC, AICc e BIC (O menor resultado é sempre o melhor)
modelo_ets = ets(AirPassengers)
modelo_ets

# Previsão
# Final do gráfico -> Linha azul = previsão | shade = intervalo de confiança
previsao_ets = forecast(modelo_ets, h=12)
previsao_ets
plot(previsao_ets)
```

## Previsões - arima
- auto.arima() -> Retorna elementos 'automáticos' para a previsão.
- autoregressão -> auto (2, 0)
- diferenciação -> auto (1, 1)
- média movel -> auto (1, 0)
- Retorno dos elementos -> ARIMA (2,1,1) -> não sazional | (0,1,0) -> sazional | [12] -> Intervalos 

```r
# Modelo ARIMA
modelo_arima = auto.arima(AirPassengers)
modelo_arima

# Previsão
previsao_arima = forecast(modelo_arima, h=12)
previsao_arima
plot(previsao_arima)
```

## Ets VS Arima
- Devido aos menores indices do modelo ARIMA, o temos como o mais indicado apra o caso...
- ets -> AIC = 1395.166
- arima -> AIC = 1017.85
