```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
data = pd.read_csv('dados/orchard.csv')
data
```

## Gráfico 3D

```python
figura = plt.figure()
eixo = axes3d.Axes3D(fig=figura)
figura.add_axes(eixo)
eixo.scatter(xs=data.decrease, ys=data.rowpos, zs=data.colpos)
eixo.set_xlabel('Decrease')
eixo.set_ylabel('Rowpos')
eixo.set_zlabel('Colpos')

# Adjust zlabel overflow!
eixo.zaxis.labelpad=-4
```
