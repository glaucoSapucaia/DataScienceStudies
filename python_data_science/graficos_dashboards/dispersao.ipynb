```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Gráficos de Dispersão

```python
plt.scatter(x=data.Girth, y=data.Volume, marker='*', color='black', facecolors='white')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

### Dispersão em Linha

```python
plt.plot(data.Girth, data.Volume, color='purple')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

## Seaborn

### Tendencia
- x_jitter -> Evita sobreposição de dados.
- fit_reg -> Cria uma linha de tendencia.

```python
sns.regplot(x=data.Girth, y=data.Volume, x_jitter=1.0, fit_reg=True)
```

### Dispesão
- hue= indica agrupamento de dados

```python
sns.scatterplot(x=data2.conc, y=data2.uptake, hue=data2.Type).set_title('Type')
```

### Filtrando dados

```python
quebec_data2 = data2.loc[data2['Type'] == 'Quebec']
missis_data2 = data2.loc[data2['Type'] == 'Mississippi']

plt.figure(1, layout='constrained')
plt.subplot(1, 2, 1)
sns.scatterplot(x=quebec_data2.conc, y=quebec_data2.uptake).set_title('Quebec')
plt.subplot(1, 2, 2)
sns.scatterplot(x=missis_data2.conc, y=missis_data2.uptake).set_title('Mississippi')
```

```python
sns.scatterplot(x=data2.conc, y=data2.uptake, hue=data2.Treatment).set_title('Treatment')
```

```python
chilled_data2 = data2.loc[data2['Treatment'] == 'chilled']
nonchil_data2 = data2.loc[data2['Treatment'] == 'nonchilled']

plt.figure(2, layout='constrained')
plt.subplot(1, 2, 1)
sns.scatterplot(x=chilled_data2.conc, y=chilled_data2.uptake).set_title('Chilled')
plt.subplot(1, 2, 2)
sns.scatterplot(x=nonchil_data2.conc, y=nonchil_data2.uptake).set_title('Nonchilled')
```

### Dispersão com catplot()
- jitter= -> Evita sobreposições

```python
sns.catplot(x='alcgp', y='ncontrols', data=data3, jitter=True)
```

### Dispersão relacionado ao tributo 'tobgp'

```python
sns.catplot(x='alcgp', y='ncontrols', col='tobgp', data=data3, jitter=True)
```
