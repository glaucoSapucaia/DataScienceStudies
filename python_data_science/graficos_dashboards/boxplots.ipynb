```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dados/trees.csv')
data.head()
```

## Boxplot

```python
plt.boxplot(data.Volume, vert=False, notch=True, showfliers=False,
            patch_artist=True)
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
plt.boxplot(data, vert=True, patch_artist=True, showfliers=True)
plt.title('Árvores')
plt.xlabel('Dados')
```
