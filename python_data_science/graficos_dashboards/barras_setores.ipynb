```python
import pandas as pd
```

```python
data = pd.read_csv('dados/insect.csv')
data
```

```python
count_data = data.groupby(['spray'])['count'].sum()
count_data
```

## Gráfico de Barras

```python
count_data.plot.bar(color=['blue', 'red', 'gray', 'purple', 'black', 'brown'])
```

## Gráfico de Setores

```python
count_data.plot.pie(legend=True)
```
