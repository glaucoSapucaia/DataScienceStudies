### Visualizando dados por ração

```python
plt.figure(1, layout='constrained')
plt.subplot(2, 3, 1)
sns.histplot(data2.loc[data2['feed'] == 'horsebean'].weight, kde=True).set_title('Horsebean')
plt.subplot(2, 3, 2)
sns.histplot(data2.loc[data2['feed'] == 'casein'].weight, kde=True).set_title('Casein')
plt.subplot(2, 3, 3)
sns.histplot(data2.loc[data2['feed'] == 'linseed'].weight, kde=True).set_title('Linseed')
plt.subplot(2, 3, 4)
sns.histplot(data2.loc[data2['feed'] == 'meatmeal'].weight, kde=True).set_title('Meatmeal')
plt.subplot(2, 3, 5)
sns.histplot(data2.loc[data2['feed'] == 'soybean'].weight, kde=True).set_title('Soybean')
plt.subplot(2, 3, 6)
sns.histplot(data2.loc[data2['feed'] == 'sunflower'].weight, kde=True).set_title('Sunflower')
```
