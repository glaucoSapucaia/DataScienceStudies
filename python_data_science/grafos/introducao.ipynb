```python
from igraph import Graph
from igraph import plot as plt
```

## Objeto IGRAPH
- D | U -> Grafo direcionado e nÃ£o direcionado
- N -> Grafo com nome
- W -> Grafo com peso em arestas
- T -> Grafo com tipo

```python
grafo1 = Graph(
    # definindo arestas
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0)
    ],
    # grafo direcionado
    directed=True,
)
# nomes dos vertices (sua prÃ³pria numeraÃ§Ã£o)
grafo1.vs['label'] = range(grafo1.vcount())
# dados do grafo 
plt(grafo1)
```

## Plot
- bbox -> DimensÃµes do grÃ¡fico

```python
plt(grafo1, bbox=(0, 0, 300, 300))
```

```python
grafo2 = Graph(
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0), (0, 3), (3, 2), (2, 1), (1, 0)
    ],
    directed=True
)
grafo2.vs['label'] = range(grafo2.vcount())
plt(grafo2, bbox=(0, 0, 400, 400))
```

```python
# grafo com LAÃO
grafo3 = Graph(
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0), (1, 1)
    ],
    directed=True
)
grafo3.vs['label'] = range(grafo3.vcount())
plt(grafo3, bbox=(0, 0, 300, 300))
```

```python
grafo4 = Graph(
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0)
    ],
    directed=True
)

# vertice isolado
grafo4.add_vertex('4')

grafo4.vs['label'] = (range(grafo4.vcount()))
print(grafo4)
plt(grafo4, bbox=(0, 0, 400, 400))
```

## Grafos nÃ£o direcionados

```python
grafo5 = Graph(directed=False)

# adicionando elementos

# multiplos vertices
grafo5.add_vertices(10)

# add um vertice
grafo5.add_vertex('16')

# add arestas
grafo5.add_edges(
    [
        (0, 1), (2, 2), (2, 3), (3, 0)
    ]
)
grafo5.vs['label'] = range(grafo5.vcount())

print(grafo5)
plt(grafo5, bbox=(0, 0, 300, 300))
```

### Personalizando nomes

```python
from os import name


grafo6 = Graph(directed=False)
grafo6.add_vertices(5,
                    attributes={
                        name: ['A', 'B', 'C', 'D', 'E']
                    })
grafo6.add_edges(
    [
        (0, 1), (1, 2), (2, 3), (3, 4), (4, 0), (0, 2), (2, 1)
    ],
)
grafo6.add_vertex('5')
grafo6.add_vertex('6')

# nomes personalizados
grafo6.vs['name'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

print(grafo6)

# SÃ³ foi possivel plotar os nomes | labels dos vertices indicando-os no plot!
plt(grafo6, vertex_label=['A', 'B', 'C', 'D', 'E', 'F', 'G'])
```

```python
grafo7 = Graph(directed=False)
grafo7.add_vertices(5)
grafo7.add_edges(
    [
        (0, 1), (1, 2), (2, 3), (3, 4), (4, 0), (0, 2), (2, 1)
    ]
)
grafo7.add_vertex('5')
grafo7.add_vertex('6')

# adicionando tanto label como name, o plot funcionou =)
grafo7.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
grafo7.vs['name'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

print(grafo7)
plt(grafo7, bbox=(0, 0, 400, 400))
```

## Matriz de adjacencia

```python
print(grafo7.get_adjacency())
```

## Buscando linhas e colunas

```python
print(grafo7.get_adjacency()[0, ])
print(grafo7.get_adjacency()[0, 0])
```

### Buscando vertices

```python
for v in grafo7.vs:
    print(v)
```

## VÃ©rtices (Nodes)

```python
grafo8 = Graph(
    edges=[
        (0, 1), (2, 3), (0, 2), (0, 3)
    ],
    directed=True
)

# attrs
grafo8.vs['label'] = ['Fernando', 'Pedro', 'Maria', 'Leticia']
grafo8.vs['idade'] = [40, 14, 27, 43]

print(grafo8)
```

```python
for v in grafo8.vs:
    print(v)

print('_' * 100)
print(grafo8.vs[0])
```

## Arestas (Edges)

```python
grafo8.es['tipoAmizade'] = ['Amigo', 'Inimigo', 'Conhecido', 'Amigo']
print(grafo8)
```

## Pesos dos Edges
Tornando um grafo PONDERADO
- weigth -> Atributo de peso do IGRAPH

```python
grafo8.es['weight'] = [1, 2, 2, 3]
print(grafo8)
```

### Visualizando arestas
