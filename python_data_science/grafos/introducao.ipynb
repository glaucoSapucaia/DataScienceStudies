```python
from igraph import Graph
from igraph import plot
```

## Objeto IGRAPH
- D -> Grafo direcionado
- N -> Grafo com nome
- W -> Grafo com peso em arestas
- B -> Grafo com tipo

```python
grafo1 = Graph(
    # definindo arestas
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0)
    ],
    # grafo direcionado
    directed=True,
)
# nomes dos vertices (sua própria numeração)
grafo1.vs['label'] = range(grafo1.vcount())
# dados do grafo 
print(grafo1)
```

## Plot
- bbox -> Dimensões do gráfico

```python
plot(grafo1, bbox=(0, 0, 300, 300))
```

```python
grafo2 = Graph(
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0), (0, 3), (3, 2), (2, 1), (1, 0)
    ],
    directed=True
)
grafo2.vs['label'] = range(grafo2.vcount())
plot(grafo2, bbox=(0, 0, 400, 400))
```

```python
# grafo com LAÇO
grafo3 = Graph(
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0), (1, 1)
    ],
    directed=True
)
grafo3.vs['label'] = range(grafo3.vcount())
plot(grafo3, bbox=(0, 0, 300, 300))
```

```python
grafo4 = Graph(
    edges=[
        (0, 1), (1, 2), (2, 3), (3, 0)
    ],
    directed=True
)

# vertice isolado
grafo4.add_vertex('4')

grafo4.vs['label'] = (range(grafo4.vcount()))
plot(grafo4, bbox=(0, 0, 400, 400))
```
