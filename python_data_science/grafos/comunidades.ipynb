```python
import igraph
from igraph import Graph, plot
import numpy as np
```

```python
grafo = igraph.load('Grafo.graphml')
print(grafo)
```

```python
plot(grafo, bbox=(0, 0, 400, 400))
```

# Comunidades

```python
comunidades = grafo.connected_components()
print(comunidades)
```

```python
# Retorna a qual cluster o elemento pertence
comunidades.membership
```

## Comunidades e cores
Diferenciando cada comunidade com o uso de cores

```python
cores = comunidades.membership
cores = np.array(cores)
cores = cores * 20
cores = cores.tolist()
plot(grafo, vertex_color = cores)
```

```python
grafo2 = Graph(
    edges=[
        (0, 2), (0, 1), (1, 4), (1, 5), (2, 3), (6, 7), (3, 7), (4, 7), (5, 6)
    ],
    directed=True
)
grafo2.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo2.es['weight'] = [2, 1, 2, 1, 2, 1, 3, 1]
```

```python
plot(grafo2, bbox=(0, 0, 400, 400), vertex_label = grafo2.vs['label'])
```
