```python
from igraph import Graph
from igraph import plot
```

```python
grafo = Graph(
    edges=[
        (0, 1), (2, 3), (0, 2), (0, 3)
    ],
    directed=True
)
# Nodes attrs
grafo.vs['label'] = ['Joaquim', 'Maria', 'Claudio', 'Manoela']
grafo.vs['idade'] = [30, 12, 15, 46]

# Edges attrs
grafo.es['Relação'] = ['Sobrinho', 'Tio', 'Pai', 'Mãe']
grafo.es['weight'] = [1, 1, 3, 3]
```

```python
for v in grafo.vs:
    print(v)

print()
print('_' * 100)
print()

for a in grafo.es:
    print(a)
```

## Visualizações

### Cores e labels

```python
grafo.vs['cor'] = ['blue', 'red', 'yellow', 'green']
plot(grafo, bbox=(0, 0, 400, 400), vertex_label=grafo.vs['label'], vertex_color=grafo.vs['cor'])
```

### Weights e idades (pesos)
- edge_curved() -> Destorce arestas, interessante para evitar a sobreposição das mesmas.
- vertex_shape() -> Indica o formato do Node, podemos utilziar imagens personalizadas.

```python
plot(
    grafo,
    bbox=(300, 300),
    vertex_color=grafo.vs['cor'],
    vertex_size=grafo.vs['idade'],
    vertex_label = grafo.vs['label'],
    edge_width=grafo.es['weight'],
    edge_label = grafo.es['Relação'],

    # mais opções gráficas
    edge_curved=0.4,
    vertex_shape='triangle'
)
```
