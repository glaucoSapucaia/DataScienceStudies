## Graus do Grafo
- in -> Graus de entrada
- out -> Graus de saida
- all -> Todos os graus

```python
graus = grafo.degree(mode='all')
print(graus)
```

```python
# Plot com representaÃ§Ã£o grÃ¡fica dos graus
plot(grafo, vertex_size=graus, bbox=(0, 0, 500, 500))
```
