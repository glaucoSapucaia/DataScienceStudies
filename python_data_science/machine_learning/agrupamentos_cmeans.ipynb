````python
import numpy as np
import skfuzzy as fuzz
from sklearn import datasets
from sklearn.metrics import confusion_matrix, accuracy_score
````

````python
iris = datasets.load_iris()
````

## Criando CLUSTERS | C-means
- T -> Faz a trasnposição de uma matriz | matriz transposta
- c= -> Qtd de clusters

````python
r = fuzz.cmeans(
    data=iris.data.T, c=3, m=2, error=0.005,
    maxiter=1000,
    init=None
)
r
````

## Retorno Percentual

````python
previsores_porcentagem = r[1]
previsores_porcentagem
````

````python
for x in range(150):
    print(
        previsores_porcentagem[0][x],
        previsores_porcentagem[1][x],
        previsores_porcentagem[2][x],
    )
````

## Perormance | Confusão
- axis=0 -> Buscamos as linhas

````python
previsoes = previsores_porcentagem.argmax(axis=0)
confusao = confusion_matrix(iris.target, previsoes)
confusao
````

````python
taxa_acerto = accuracy_score(iris.target, previsoes)
taxa_acerto
````
