## K-Means
Agrupamento simples que utiliza a distância euclidiana.  
Precisamos definir o número de CLUSTERS

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
iris
```

## Visualização da quantidade de classes e registros de cada uma delas

```python
unicos, quantidade = np.unique(iris.target, return_counts=True)
unicos
```

```python
quantidade
```

## Agrupamento K-Means | Tarefa Não Supervisionada
- n_clusters=3 -> Indica a quantidades de nodes (classes | grupos)

## init= e random_state=
- Caso perceba alguma inconsistencia nos agrupamentos, execute o código novamente

```python
cluster = KMeans(n_clusters=3, init='k-means++')
cluster.fit(iris.data)
```

## Visualizando centroides (centro dos agrupamentos)

```python
centroides = cluster.cluster_centers_
centroides
```

## Visualizando Agrupamentos

```python
agrupamentos = cluster.labels_
agrupamentos
```

## Visualização da quantidade de classes e registros de cada uma delas

```python
unicos, quantidade = np.unique(agrupamentos, return_counts=True)
unicos
```

```python
quantidade
```

## Performance | Confusão

```python
resultados = confusion_matrix(iris.target, agrupamentos)
resultados
```

## Visualização Gráfica

```python
plt.scatter(
    iris.data[agrupamentos == 0, 0],
    iris.data[agrupamentos == 0, 1],
    c='green',
    label='Setosa'
)
plt.scatter(
    iris.data[agrupamentos == 1, 0],
    iris.data[agrupamentos == 1, 1],
    c='red',
    label='Versicolor'
)
plt.scatter(
    iris.data[agrupamentos == 2, 0],
    iris.data[agrupamentos == 2, 1],
    c='blue',
    label='Virgica'
)
plt.legend()
```
