```python
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm, skewnorm
```

## Distribuição Normal
- norm.rvs() -> Cria dados com distribuição normal

```python
data = norm.rvs(size=1000)
data
```

```python
plt.hist(data, bins=20)
plt.title('Dados')
```

## Verificando Distribuição Normal | QQ Plot

```python
fig, ax = plt.subplots()
stats.probplot(data, fit=True, plot=ax)
plt.show()
```

## Teste Shapiro
- pvalue= -> Se acima de 0.05, não rejeitamos hipótese nula | É Normalmente Distribuido

```python
stats.shapiro(data)
```

## Dados Não Normais
- skewnorm.rvs() -> Gera dados não normalmente distribuidos

```python
data2 = skewnorm.rvs(4, size=1000)
data2
```

```python
plt.hist(data2, bins=20)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(data2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(data2)
```
