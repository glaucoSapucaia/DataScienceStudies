```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import numpy as np
```

## Separando elementos

```python
data_ = datasets.load_iris()
previsores = data_.data
previsores
```

### As classes já estão num formato numérico!

```python
classes = data_.target
classes
```

## Transformação de dados
Transformamos as classes para o formato "dummy", pois temos uma rede neural com 3 saidas.

```python
classe_dummy = to_categorical(classes)
classe_dummy
```

# Treino | Teste

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(
    previsores,
    classe_dummy,
    test_size=0.3,
    random_state=0
)
```

# Rede Neural
- Estrutura com sequencia de camadas (Sequential)
- Dense() -> Gera camadas 100% conectadas (todos os neuronios conectados)
- units= -> Qtd de neuronios
- input_dim -> Neuronios de entrada (4 variaveis independentes de Iris)
- activation -> funcao de ativação | softmax (gera probabilidades para cada saida)  
  
Como são três classes, não podemos utilizar uma função de ativação simples com retorno de 0 ou 1 por exemplo.

```python
modelo = Sequential()

# camada oculta 1 + camada de entrada
modelo.add(Dense(units=5, input_dim=4))

# camada oculta 2
modelo.add(Dense(units=4))

# camada oculta 3 e função de ativação
modelo.add(Dense(units=3, activation='softmax'))

modelo.summary()
```

## Compilação | Treinamento
- compile() -> Parametros do modelo
- optimizer -> Otimiza os pesos
- loss -> Calcula a diferença entre valores previstos e reais, e realiza calculo para juste dos pesos
- metrics -> Analisa performance
- epochs -> Qtd de testes da rede neural
- validation_data -> Dados de teste

```python
# Parametros da rede neural
modelo.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Treinamento da rede neural
modelo.fit(
    x_treino,
    y_treino,
    epochs=1000,
    validation_data=(x_teste, y_teste)
)
```

## Previsoes | Confusão

```python
previsoes = modelo.predict(x_teste)

# Adaptação para o retorno em true | false
previsoes = (previsoes > 0.5)
previsoes
```

```python
# Buscando maior valor de cada saida
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

```python
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
