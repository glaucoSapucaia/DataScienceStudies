# Dispersão
- Excelente para comparação de 2 variaveis numéricas. Possível observar tendencias!

### matplot

```python
plt.scatter(x=trees.Girth, y=trees.Volume, color='purple', facecolors='none', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
plt.plot(trees.Girth, trees.Volume)
plt.title('Árvores')
plt.ylabel('Volume')
plt.xlabel('Circunferência')
```

### seaborn
- attr x_jitter controla a sobreposição de dados!
- attr fir_reg cria linha de tendencia.

```python
sns.regplot(x=trees.Girth, y=trees.Volume, data=trees, x_jitter=0.3, fit_reg=True)
```

## Dispersão com legendas

```python
DB_PATH_CO2 = 'C:\\Users\\glauc\\OneDrive\\Área de Trabalho\\data_science\\studies\\FormacaoCD\\13.Prática em Python\\dados\\co2.csv'
co2 = pd.read_csv(DB_PATH_CO2)
co2.head()
```

```python
conc_ = co2.conc
uptake_ = co2.uptake
treatments_ = list(set(co2.Treatment))
treatments_
```

### Gerando legendas!

```python
for i in range(len(treatments_)):
    indice = co2.Treatment == treatments_[i]
    plt.scatter(conc_[indice], uptake_[indice], label=treatments_[i])
plt.legend(loc='lower right')
```
