# Dispersão

```python
DB_PATH_CO2 = 'C:\\Users\\glauc\\OneDrive\\Área de Trabalho\\data_science\\studies\\FormacaoCD\\13.Prática em Python\\dados\\co2.csv'
co2 = pd.read_csv(DB_PATH_CO2)
co2
```

```python
srn.scatterplot(x=co2.conc, y=co2.uptake, hue=co2.Type)
```

## Separando dados pela coluna "Type"

```python
quebec_ = co2.loc[co2['Type'] == 'Quebec']
mississippi_ = co2.loc[co2['Type'] == 'Mississippi']
plt.figure(1)
plt.subplot(1, 2, 1)
srn.scatterplot(x=quebec_.conc, y=quebec_.uptake).set_title('Quebec')
plt.subplot(1, 2, 2)
srn.scatterplot(x=mississippi_.conc, y=mississippi_.uptake).set_title('Mississippi')
plt.tight_layout()
```

## Separando dados pela coluna "Treatment"

```python
nonchilled_ = co2.loc[co2['Treatment'] == 'nonchilled']
chilled_ = co2.loc[co2['Treatment'] == 'chilled']
plt.figure(2)
plt.subplot(1, 2, 1)
srn.scatterplot(x=chilled_.conc, y=chilled_.uptake).set_title('Chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(x=nonchilled_.conc, y=nonchilled_.uptake).set_title('Non-chilled')
plt.tight_layout()
```
