```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DB_PATH_TREES = 'C:\\Users\\glauc\\OneDrive\\Área de Trabalho\\data_science\\studies\\FormacaoCD\\13.Prática em Python\\dados\\trees.csv'
trees = pd.read_csv(DB_PATH_TREES)
trees.head()
```

# Dashboars

## Girth and Volume

```python
plt.scatter(trees.Girth, trees.Volume, color='red')
```

## Girth and Height

```python
plt.scatter(trees.Girth, trees.Height, color='purple')
```

## Height and Volume

```python
plt.scatter(trees.Height, trees.Volume, marker='*')
```

## Histogram - Volume

```python
plt.hist(trees.Volume, bins=6, color='red')
```

# All plots!

```python
plt.figure(1)
plt.subplot(2, 2, 1)
plt.scatter(trees.Girth, trees.Volume, color='red')
plt.title('Circunferencia x Volume')
plt.subplot(2, 2, 2)
plt.scatter(trees.Girth, trees.Height, color='purple')
plt.title('Circunferencia x Altura')
plt.subplot(2, 2, 3)
plt.scatter(trees.Height, trees.Volume, marker='*')
plt.title('Altura x Volume')
plt.subplot(2, 2, 4)
plt.hist(trees.Volume, bins=6, color='red')
plt.title('Volume')
```

# Boxplot

```python
plt.boxplot(trees.Volume, vert=False, showfliers=False, notch=True, patch_artist=True)
```

### Todos os 3 dados numéricos

```python
plt.boxplot(trees, showfliers=False)
plt.title('Árvores')
plt.xlabel('Dados')
```
