## Db = dados.csv

```r
library(lattice)
library(RColorBrewer)
data = read.csv('dados.csv', sep=';', na.strings='', stringsAsFactors=T)
summary(data)
head(data)
```

## Valores nulos ou duplicados

```r
data[!complete.cases(data), ]
data[duplicated(data$CODIGO), ]
data = data[-c(20, 22, 23, 43, 58, 59, 84, 95, 96, 127, 142, 143, 181, 182, 183), ]
data[duplicated(data$CODIGO), ]
```

## PIB

```r
summary(data$PIB)
bwplot(data$PIB, main='Municipios', xlab='PIB')
media = median(data$PIB)
desv = sd(data$PIB)
data[data$PIB >= 2 * desv, ]$PIB = media
summary(data$PIB)
bwplot(data$PIB, main='Municipos', xlab='PIB')
histogram(data$PIB, main='Municipios', xlab='PIB', nint=6, col=c(1:6),
type='count', aspect=1)
boxplot(data$PIB, main='Municipios', xlab='PIB', outline=F, col='purple',
horizontal=T)
```

## Valor Empenho

```r
summary(data$VALOREMPENHO)
bwplot(data$VALOREMPENHO, main='Municipios', xlab='Valor Empenho')
media = median(data$VALOREMPENHO)
desv = sd(data$VALOREMPENHO)
data[data$VALOREMPENHO >= 2 * desv, ]$VALOREMPENHO = media
summary(data$VALOREMPENHO)
bwplot(data$VALOREMPENHO, main='Municipios', xlab='Valor Empenho', col='purple')
histogram(data$VALOREMPENHO, main='Municipios', xlab='Valor Empenho',
col=c(1:5), nint=5, aspect=1, type='percent')
boxplot(data$VALOREMPENHO, main='Municipios', xlab='Valor Empenho', col='blue',
horizontal=T, outline=F)
```

## Maiores PIBS

```r
maior_pib = head(data[order(-data$PIB), ], 5)
maior_pib
barplot(maior_pib$PIB, col=c(1:5), las=2,
main='Maior PIB')
legend(x='bottomright', legend=maior_pib$MUNICIPIO, col=c(1:5),
lty=1:2, cex=1, ncol=1, lwd=10)
box()
```

## Maior valor empenho

```r
maior_valor = head(data[order(-data$VALOREMPENHO), ], 5)
maior_valor
barplot(maior_valor$VALOREMPENHO, col=c(1:5), main='Maior VALOR EMPENHO')
legend(x='bottomright', legend=maior_valor$MUNICIPIO, col=c(1:5),
lty=1:2, cex=1, lwd=10)
```

## Maior proporção PIB e Valor Empenho

```r
data$PROPORCAO = data$VALOREMPENHO / data$PIB
maior_proporcao = head(data[order(-data$PROPORCAO), ], 5)
barplot(maior_proporcao$PROPORCAO, main='Maior VALOR EMPENHO x PIB', col=c(1:5), )
legend(x='bottomright', col=c(1:5), legend=maior_proporcao$MUNICIPIO, cex=1,
lty=1:2, lwd=10)
```

## Gráficos condicionais


```r
xyplot(PIB ~ VALOREMPENHO, data=data, pch=19)
densityplot(PIB ~ VALOREMPENHO, data=data, plot.points=F)
```

## Entrega
