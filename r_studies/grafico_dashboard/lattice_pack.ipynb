```r
summary(chickwts)
histogram(chickwts$weight)
```

### Agregando pesos pelo dado de alimentação!

```r
aggregate(chickwts$weight, by=list(chickwts$feed), FUN = sum)
histogram(~weight | feed, data=chickwts)
```
